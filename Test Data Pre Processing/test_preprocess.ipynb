```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
import re
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn import metrics
import pylab as pl
import unidecode
import pickle

wordnet_lemmatizer = WordNetLemmatizer()
```

```python
filename= ''
dataset = pd.read_csv(filename,encoding='utf-8-sig',header=None)
X=dataset.iloc[:,6].values
X=pd.Series(X)

wordnet_lemmatizer = WordNetLemmatizer()

def stem(tweet):
    stemmer = nltk.stem.PorterStemmer()
    tweet_stem = ''
    words = [word if(word[0:2]=='__') else word.lower() \
                for word in tweet.split() \
                if len(word) >= 3]
    words = [stemmer.stem(w) for w in words] 
    tweet_stem = ' '.join(words)
    return tweet_stem

    
#lemmatization

def lemma(tweet):
    punctuations="?:!.,;"
    sentence_words = nltk.word_tokenize(tweet)
    sentence_words_lemma = []
    for word in sentence_words:
        if word in punctuations:
            sentence_words.remove(word)
        else:
            sentence_words_lemma.append(wordnet_lemmatizer.lemmatize(word,pos="v"))
    return ' '.join(sentence_words_lemma)

#stop words removal
swords = set(stopwords.words('english'))
def no_user_alpha(tweet):
    tweet_list = [ele for ele in tweet.split() if ele != 'user']
    clean_tokens = [t for t in tweet_list if re.match(r'[^\W\d]*$', t)]
    clean_s = ' '.join(clean_tokens)
    clean_mess = [word for word in clean_s.split() if word.lower() not in swords]
    return ' '.join(clean_mess)
```

```python
def preprocess(tweet):
    
    #Convert www.* or https?://* to URL
    tweet = re.sub('((www\.[^\s]+)|(https?://[^\s]+))','URL',tweet)
    
    #Convert @username to __USERHANDLE
    tweet = re.sub('@[^\s]+','__USERHANDLE',tweet)  
    
    #Replace #word with word
    tweet = re.sub(r'#([^\s]+)', r'\1', tweet)
    
    #trim
    tweet = tweet.strip('\'"')
    
    # Repeating words like hellloooo
    repeat_char = re.compile(r"(.)\1{1,}", re.IGNORECASE)
    tweet = repeat_char.sub(r"\1\1", tweet)
    
    #remove non ascii
    tweet = unidecode.unidecode(tweet)
    

     #Convert to lower case
    tweet = tweet.lower()
    
    return tweet



#Use different methods to fine tune your algorithm

X = [lemma(preprocess(tweet)) for tweet in X]
#X = [stem(preprocess(tweet)) for tweet in X]
#X = [no_user_alpha(preprocess(tweet)) for tweet in X]
#X = [preprocess(tweet) for tweet in X]
```
